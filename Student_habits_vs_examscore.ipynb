Importing the required library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

Loading the dataset 

```python
df=pd.read_csv("student_habits_performance.csv" )
df
```

Removing the unecessary columns which are not important for the analysis 

```python
df1=df.drop(columns=["netflix_hours","part_time_job","diet_quality","exercise_frequency",
                     "parental_education_level","internet_quality","extracurricular_participation"])
df1
```

Checking for any null values in the dataset 

```python
df1.isnull().sum()
```

```python
df1.dtypes
```

```python
df1.set_index('student_id',inplace=True)
df1
```

Saving the cleaned data 

```python
df1.to_csv('cleaned_data.csv',index=True)
```

# EDA 

```python
df1.describe()
```

Creating a new column where if exam_score was less than 40 it was fail , if it was less than 80 it was Average and above 80 is considered as good .

```python
def evaluate_result(marks):
    if marks >= 80:
        return 'Good'
    elif marks > 40:
        return 'Average'
    else:
        return 'Fail'

df1['student_performance'] = df1['exam_score'].apply(evaluate_result)
df1
```

The graph of the exam score is is left skewed , that shows most of the students scored good marks and fewer students got good marks .

```python

sns.histplot(df1['exam_score'], kde=True)

# Add vertical lines for mean and quartiles
plt.axvline(x=np.mean(df1['exam_score']), color='red', linestyle="--", label='Mean')
plt.axvline(x=np.percentile(df1['exam_score'], 25), color='green', linestyle="--", label='1st Quartile (25%)')
plt.axvline(x=np.percentile(df1['exam_score'], 75), color='black', linestyle="--", label='3rd Quartile (75%)')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

Now we will find out the variables that are correlated with exam score using correlation matrix. 

```python
df1.select_dtypes(include=["number"]).corr()
```

From the above matrix we can infer :
1)Age has rarely any effect on exam score  .
2)Study hours has strong correlation with exam score.
3)Social media hours has a slight  negative correlation with exam score.
4)Attendance percentage also has a slight positive correlation with exam score .
5)Sleep hours also have a slight positive correlation with exam score.
6)Metal health rating  has a slight positive correlation with exam score.
